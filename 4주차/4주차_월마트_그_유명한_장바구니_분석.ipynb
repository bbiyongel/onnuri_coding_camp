```python
import pandas as pd

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('OnlineRetail.xlsx')
```

```python
basket = (df.groupby(['InvoiceNo','Description'])['Quantity'].sum()
          .unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
df1 = df.groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
df1.head(20)
```

```python
df2 = df1.unstack()
```

```python
df3 = df2.reset_index()
```

```python
df3.head()
```

```python
df4 = df3.fillna(0)
```

```python
df4.head()
```

```python
df5 = df4.set_index('InvoiceNo')
```

```python
df5.head()
```

```python
def enc_func(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
```

```python
bsk = basket.applymap(enc_func)
```

```python
bsk.head()
```

```python
freq_item = apriori(bsk, min_support=0.07, use_colnames=True)
```

```python
freq_item
```

```python
rules = association_rules(freq_item, metric='lift', min_threshold=1)
```

```python
rules
```

```python
rules[(rules['lift']>=5)&(rules['confidence']>=0.8) ]
```
